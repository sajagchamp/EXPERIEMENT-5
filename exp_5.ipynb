```python
#changing/adding key value in dictionary
car={
    "brand":"ford",
    "model":"mustang",
    "year":1964
}
car["color"]="red"# or we can use car.update({"color":red})
print(car)
x=car.values()
print(x)#after the change only values not the keys
```

```python
#retrival of values
student={
    "roll":101,
    "name":"Sajag",
    "branch":"E&TC"
}
print("Name:",student["name"])
print("roll number:",student["roll"])
print("branch:",student["branch"])
```

```python
#removal of values
student={
    "roll":101,
    "name":"Sajag",
    "branch":"E&TC"
}
student.pop("branch")
print(student)
```

```python
#program to get student marks and get error if no student found
marks={"sajag":90,"ram":95,"shiv":97}
name=input("enter student name:")
print(marks.get(name,"student not found"))#pahel yey name par jayega agar name nahi hua than 'student not found' par jayega
```

```python
#program to get student marks and get error if no student found
marks={"sajag":90,"ram":95,"shiv":97}
name=input("enter student name:")
if name in marks:
    print(marks[name])
else:
    print("student not found")

```

```python
#"program to validate user login"
user={
    "sajag":"sajag123",
    "ram":"ram123"
}
username=input("enter username:")
password=input("enter password:")
if username in user and user[username]==password:
    print("login successful")
else:
    print("invalid username or password")
```

```python
#program to find student who scored highest marks,(limitation:student with equal marks needs to be updated than)
marks={"A":85,"B":98,"C":76}
topper=max(marks,key=marks.get)#agar ham key=marks.get kar rhe hai to hame a b c mai sai jo highest hoga voh lega so key=marks.get matlab ham ab uski values ki baat kar rhe hai
print("topper:",topper)
print("marks:",marks[topper])

```
